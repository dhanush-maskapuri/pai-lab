```python

from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination
model = DiscreteBayesianNetwork([('A', 'B'), ('B', 'C')])
cpd_A = TabularCPD(variable='A', variable_card=2, values=[[0.6], [0.4]])
cpd_B = TabularCPD(variable='B', variable_card=2,
                   values=[[0.7, 0.1], [0.3, 0.9]],
                   evidence=['A'],
                   evidence_card=[2])
cpd_C = TabularCPD(variable='C', variable_card=2,
                   values=[[0.8, 0.4], [0.2, 0.6]],
                   evidence=['B'],
                   evidence_card=[2])
model.add_cpds(cpd_A, cpd_B, cpd_C)
assert model.check_model()
inference = VariableElimination(model)
result = inference.query(variables=['C'], evidence={'A': 1})
print("Inference Result:")
print(result)
result_b_given = inference.query(variables=['C'], evidence={'B': 0})
print("\nInference Result (C | B=0):")
print(result_b_given)
```
